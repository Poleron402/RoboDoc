## Can a model be trained to benefit predictions?


```python
# Needed methods

def rmse(predicted, actual):
    return np.sqrt(((predicted - actual)**2).mean())

def get_best(ks, rmse):
    smallest = np.min(rmse)
    for i in range(len(rmse)):
        if rmse[i] == smallest:
            return ks[i], rmse[i]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

label_encoder = LabelEncoder()  #Encoding the Disease names 
df_learning = df.copy()
df_learning["Disease_encode"] = label_encoder.fit_transform(df["Disease"])



X = df_learning.drop(columns=["Disease", "Disease_encode"]).values
y = df_learning['Disease_encode'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = KNeighborsClassifier(n_neighbors=1, algorithm='brute')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
y_pred = y_pred.astype(int)
y_pred_names = label_encoder.inverse_transform(y_pred)
y_test_names = label_encoder.inverse_transform(y_test)

# Display actual vs. predicted
results_df = pd.DataFrame({"Actual Disease": y_test_names, "Predicted Disease": y_pred_names})
print(results_df.head())
```

Getting the best value of k based on RMSE

```python

n = 30
test_rmse = []
ks = np.arange(1, n+1)
for k in ks:
    regr = KNeighborsClassifier(n_neighbors=k, algorithm='brute')
    model = regr.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    rmse_te = rmse(y_train, y_train_pred)
    test_rmse.append(rmse_te)

sns.barplot(x = ks, y =test_rmse)
```

```python
best_k, best_rmse = get_best(ks, test_rmse)
print('best k = {}, best test RMSE: {:0.1f}'.format(best_k, best_rmse))
```

###  Generating random test data for classification

```python
np.random.seed(42)
num_samples = 10

data = {
    "Glucose": np.random.rand(num_samples),
    "Cholesterol": np.random.rand(num_samples),
    "Mean Corpuscular Hemoglobin": np.random.rand(num_samples),
    "Platelets": np.random.rand(num_samples),
    "White Blood Cells": np.random.rand(num_samples),
    "Disease": np.random.choice(["Healthy", "Diabetes", "Anemia", "Thalassemia"], num_samples)
}

df_test = pd.DataFrame(data)

# Introduce some missing values (NAs)
df_test.loc[2, "Glucose"] = np.nan
df_test.loc[5, "Cholesterol"] = np.nan
df_test.loc[7, "Mean Corpuscular Hemoglobin"] = np.nan
```
